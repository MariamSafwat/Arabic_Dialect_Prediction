```python
import pandas as pd
import seaborn as sns 
import eli5
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
from sklearn import metrics
from sklearn.model_selection import train_test_split
import pickle
```

```python
train_df = pd.read_csv('/content/drive/MyDrive/Dialect_Identification/train.csv')
test_df = pd.read_csv('/content/drive/MyDrive/Dialect_Identification/test.csv')
```

```python
def print_report(pipe, x_test, y_test):
    y_pred = pipe.predict(x_test)
    report = metrics.classification_report(y_test, y_pred)
    print(report)
    print("accuracy: {:0.3f}".format(metrics.accuracy_score(y_test, y_pred)))
```

# ML Model

---



Logistic Regression

```python
vec = CountVectorizer()
clf = LogisticRegression()
pipe = make_pipeline(vec, clf)
pipe.fit(train_df.text, train_df.dialect)
```

Test the model

```python
print_report(pipe, test_df.text, test_df.dialect)
```

```python
eli5.show_weights(pipe, vec=vec, top=20)
```

Save the model

```python
with open('/content/drive/MyDrive/Dialect_Identification/LR_model.pkl','wb') as f:
    pickle.dump(pipe,f)
```

## **LSTM**

```python
import numpy as np 

import tensorflow as tf
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Dense, Embedding, LSTM, SpatialDropout1D
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Dialect_Identification/df_final.csv')
```

Get length of longest tweet 

```python
def get_longest_text(texts):
    longest_input = 0
    for text in texts:
        text_len= len(text.split())
        longest_input = max(longest_input, text_len)
    return longest_input
```

```python
longest_input = get_longest_text(df['text'])
longest_input
```

```python
# The maximum number of words to be used. (most frequent)
MAX_NB_WORDS = 50000
# This is fixed.
EMBEDDING_DIM = 100
tokenizer = Tokenizer(num_words = MAX_NB_WORDS, lower = True)
tokenizer.fit_on_texts(df['text'].values)
word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))
```

Pad all tweets with maximum length of the longest one

```python
X = tokenizer.texts_to_sequences(df['text'].values)
X = pad_sequences(X, maxlen=longest_input)
print('Shape of data tensor:', X.shape)
```

```python
Y = pd.get_dummies(df['dialect']).values
print('Shape of label tensor:', Y.shape)
```

**Train test split**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.10, random_state = 42, shuffle=True, stratify=Y)
print(X_train.shape,Y_train.shape)
print(X_test.shape,Y_test.shape)
```

**Build the model**

```python
model = Sequential()
model.add(Embedding(MAX_NB_WORDS, EMBEDDING_DIM, input_length=X.shape[1]))
model.add(SpatialDropout1D(0.2))
model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(18, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
history = model.fit(X_train, Y_train, validation_split=0.1)
```

```python
accr = model.evaluate(X_test,Y_test)
print('Test set\n  Loss: {:0.3f}\n  Accuracy: {:0.3f}'.format(accr[0],accr[1]))
```
